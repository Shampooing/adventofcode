```python
with open('input', 'r') as ifile:
    pattern = [line.strip() for line in ifile]
```

```python
pattern
```

# Part 1

```python
width = len(pattern[0])
```

```python
def num_trees(rows, start_pos):
    if not rows:
        return 0
    return (1 if rows[0][start_pos] == '#' else 0) + num_trees(rows[1:], (start_pos + 3) % width)
```

```python
num_trees(pattern, 0)
```

# Part 2

```python
def num_trees_for_slope(rows, start_pos, right_move, down_move):
    if not rows:
        return 0
    return ((1 if rows[0][start_pos] == '#' else 0)
            + num_trees_for_slope(rows[down_move:], (start_pos + right_move) % width, right_move, down_move))
```

```python
[num_trees_for_slope(pattern, 0, right, down) for (right, down) in [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]]
```
